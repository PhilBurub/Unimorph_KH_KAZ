```python
def writeerror(text):
    with open('errors.txt', 'a', encoding='utf-8') as err:
        err.write(text + '\n')

def writefinal(row):
    with open('final.tsv', 'a', encoding='utf-8') as fin:
        fin.write(row)
```

```python
with open('manual check.tsv', 'r', encoding='utf-8') as f:
    for row in f.readlines():
        if row.split('\t')[3].strip() == 'TRUE':
            w_form = row.split('\t')[1]
            morphemes = row.split('\t')[2]
            if morphemes.split(';')[0] == 'N':
                noun = 0
                poss = 0
                case = 0
                num = 0
                for morph in morphemes.split(';'):
                    if morph == 'N':
                        noun += 1
                    elif morph in ('SG', 'DU', 'PL'):
                        num += 1
                    elif morph.startswith('PSS'):
                        poss += 1
                    elif morph in ('NOM', 'DAT', 'LOC'):
                        case += 1
                    else:
                        writeerror(f"MORPH_ERROR: Morpheme {morph} in wordform {w_form}")
                        continue
                if noun == 1 and num == 1 and case == 1 and poss <= 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
                    
            elif morphemes.split(';')[0] == 'PRO':
                pro = 0
                case = 0
                num = 0
                for morph in morphemes.split(';'):
                    if morph == 'PRO':
                        pro += 1
                    elif morph in ('SG', 'DU', 'PL'):
                        num += 1
                    elif morph in ('NOM', 'DAT', 'LOC'):
                        case += 1
                    else:
                        writeerror(f"MORPH_ERROR: Morpheme {morph} in wordform {w_form}")
                        continue
                if pro == 1 and num == 1 and case == 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
            
            elif morphemes.split(';')[0] == 'V':
                if 'IMP' in morphemes.split(';'):
                    pers_2 = 0
                    num = 0
                    verb = 0
                    ac = 0
                    imp = 0
                    for morph in morphemes.split(';'):
                        if morph == 'V':
                            verb += 1
                        elif morph == 'IMP':
                            imp += 1
                        elif morph == 'ARGNO2':
                            pers_2 += 1
                        elif morph in ('ARGNOS', 'ARGNOP', 'ARGNOD', 'ARGNO{D/P}'):
                            num += 1
                        elif morph in ('ARGACS', 'ARGAC', 'ARGAC{D/P}'):
                            ac += 1
                        else:
                            writeerror(f"MORPH_ERROR: Morpheme {morph} in wordform {w_form}")
                            continue
                    if imp == 1 and verb == 1 and pers_2 == 1 and num == 1 and ac <= 1:
                        writefinal(row)
                    else:
                        writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                        continue
                else:
                    verb = 0
                    tense = 0
                    spers = 0
                    snum = 0
                    anum = 0
                    infer = 0
                    taxis = 0
                    pasv = 0
                    for morph in morphemes.split(';'):
                        if morph == 'V':
                            verb += 1
                        elif morph in ('PST', 'PRS'):
                            tense += 1
                        elif morph in ('ARGNOS', 'ARGNOP', 'ARGNOD'):
                            snum += 1
                        elif morph in ('ARGNO1', 'ARGNO2', 'ARGNO3', 'ARGNO{2/3}'):
                            spers += 1
                        elif morph in ('ARGACS', 'ARGAC', 'ARGAC{D/P}'):
                            anum += 1
                        elif morph == 'INFER':
                            infer += 1
                        elif morph == 'TAXIS':
                            taxis += 1
                        elif morph == 'PASS':
                            pasv += 1
                        else:
                            writeerror(f"MORPH_ERROR: Morpheme {morph} in wordform {w_form}")
                            continue
                    if verb == 1 and tense == 1 and spers == 1 and snum == 1 and anum <= 1 and infer <= 1 and taxis <= 1 and \
                    pasv <= 1:
                        writefinal(row)
                    else:
                        writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                        continue
                        
            elif morphemes.split(';')[0] == 'ADJ':
                adj = 0
                comp = 0
                dat = 0
                for morph in morphemes.split(';'):
                    if morph == 'ADJ':
                        adj += 1
                    elif morph == 'DAT':
                        dat += 1
                    elif morph == 'COMP':
                        comp += 1
                    else:
                        writeerror(f"MORPH_ERROR: Morpheme {morph} in wordform {w_form}")
                        continue
                if adj == 1 and comp <= 1 and dat <= 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
                    
            elif morphemes.split(';')[0] == 'ADV':
                adv = 0
                comp = 0
                for morph in morphemes.split(';'):
                    if morph == 'ADV':
                        adv += 1
                    elif morph == 'COMP':
                        comp += 1
                    else:
                        writeerror(f"MORPH_ERROR: Morpheme {morph} in wordform {w_form}")
                        continue
                if adv == 1 and comp <= 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
                    
            elif morphemes.split(';')[0] == 'V.PTCP':
                verb = 0
                tense = 0
                for morph in morphemes.split(';'):
                    if morph == 'V.PTCP':
                        verb += 1
                    elif morph in ('PST', 'PRS'):
                        tense += 1
                    else:
                        writeerror(f"MORPH_ERROR: Morpheme {morph} in wordform {w_form}")
                        continue
                if verb == 1 and tense == 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
                
            elif morphemes.split(';')[0] == 'V.CVB':
                if len(morphemes.split(';')) == 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
                
            elif morphemes.split(';')[0] == 'PART':
                if len(morphemes.split(';')) == 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
                    
            elif morphemes.split(';')[0] == 'ADP':
                adp = 0
                case = 0
                poss = 0
                for morph in morphemes.split(';'):
                    if morph == 'ADP':
                        adp += 1
                    elif morph in ('LOC', 'DAT'):
                        case += 1
                    elif morph.startswith('PSS'):
                        poss += 1
                    else:
                        writeerror(f"MORPH_ERROR: Morpheme {morph} in wordform {w_form}")
                        continue
                if adp == 1 and case <= 1 and poss <= 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
                
            elif morphemes.split(';')[0] == 'INTJ':
                if len(morphemes.split(';')) == 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
                    
            elif morphemes.split(';')[0] == 'CONJ':
                if len(morphemes.split(';')) == 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
            
            elif morphemes.split(';')[0] == 'NUM':
                num = 0
                comp = 0
                for morph in morphemes.split(';'):
                    if morph == 'NUM':
                        num += 1
                    elif morph == 'COMP':
                        comp += 1
                    else:
                        writeerror(f"MORPH_ERROR: Morpheme {morph} in wordform {w_form}")
                        continue
                if num == 1 and comp <= 1:
                    writefinal(row)
                else:
                    writeerror(f"FORM_ERROR: Morphemes {morphemes} in wordform {w_form}")
                    continue
            
            else:
                writeerror(f"POS_ERROR: Morphemes {morphemes} in wordform {w_form}")
                continue
```
